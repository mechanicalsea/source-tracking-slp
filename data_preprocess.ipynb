## Classifition

```python
from sklearn.model_selection import cross_val_predict
from sklearn.svm import SVC
from sklearn.utils import shuffle
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
FEATURES = [
    'duration',
    'src_bytes',
    'dst_bytes',
    'land',
    'wrong_fragment',
    'urgent',
    'count',
    'serror_rate',
    'rerror_rate',
    'same_srv_rate',
    'diff_srv_rate',
    'srv_count',
    'srv_serror_rate',
    'srv_rerror_rate',
    'srv_diff_host_rate'
]
```

```python
features = [fe for fe in FEATURES]
```

```python
df_temp = df_normal.append(df_neptune)
```

```python
df_temp[features].head()
```

```python
X, y = df_temp[features].values, df_temp['label'].map({'normal':1, 'neptune':0}).values
X, y = shuffle(X, y)
n_sample = 5000
X, y = X[:n_sample], y[:n_sample]
```

```python
clf = SVC(C=10., kernel='rbf', random_state=42, verbose=True)
predicted = cross_val_predict(clf, X, y, cv=3)
```

```python
print confusion_matrix(y, predicted)
```

```python
X, y = df_temp[features].values, df_temp['label'].map({'normal':1, 'neptune':0}).values
```

```python
n_sample = 5000
n_predict = 20000
for ite in range(5):
    X, y = shuffle(X, y)
    Xi, yi = X[:n_sample], y[:n_sample]
    clf = SVC(C=10., kernel='rbf', random_state=42)
    clf.fit(Xi, yi)
    
    X, y = shuffle(X, y)
    Xi, yi = X[:n_predict], y[:n_predict]
    predicted = clf.predict(Xi)
    
    print 'Epoch = %d' % ite
    print confusion_matrix(yi, predicted)
    print f1_score(yi, predicted)
```
